```python
import pandas as pd
import numpy as np
```

```python
file = 'purchase_data.json'
```

```python
# Read with Pandas
purchase_df = pd.read_json(file)
purchase_df.head()
```

```python
player_count = purchase_df.loc[:, ["Gender", "SN", "Age"]]
count = player_count['SN'].nunique()
total_count = pd.DataFrame({"Total Count":[count]})
total_count
```

```python
#Count of Unique Items
unique = purchase_df['Item ID'].value_counts()
count_unique = len(unique)
#Average Price
avg_price = purchase_df['Price'].mean()
#Number of Purchase
count_purchases = len(purchase_df)
#Total Revenue
revenue = purchase_df['Price'].sum()
#Mapping
#file_df["avg_cost"] = file_df["avg_cost"].map("${:.2f}".format)


#Summary Table
summary_df = pd.DataFrame({"Number of Unique Items": count_unique, "Average Price": [avg_price],
                           "Number of Purchases": [count_purchases],
                           "Total Revenue": [revenue]})

summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

```python
#Get value counts by gender divide/total count *100
#df['column'].value_counts()
counts_by_gender = purchase_df['Gender'].value_counts()
percentage_by_gender = (100 * counts_by_gender / counts_by_gender.sum())

gender_df = pd.DataFrame({'Percentage of Players' : percentage_by_gender,'Total Count': counts_by_gender})
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2f}%".format)
gender_df
```

```python
grouped_purchases_df = purchase_df.groupby(['Gender'])

pv = grouped_purchases_df['Price'].sum()
pc = grouped_purchases_df['Price'].count()
ap = grouped_purchases_df['Price'].mean()
nt = pv / gender_df['Total Count']
purchasing_analysis_table = pd.DataFrame({'Purchase Count': pc, 'Average Purchase Price': ap, 'Total Purchase Value': pv, 'Normalized Totals': nt})

purchasing_analysis_table['Average Purchase Price'] = purchasing_analysis_table['Average Purchase Price'].map("${:,.2f}".format)
purchasing_analysis_table['Normalized Totals'] = purchasing_analysis_table['Normalized Totals'].map("${:,.2f}".format)
purchasing_analysis_table['Total Purchase Value'] = purchasing_analysis_table['Total Purchase Value'].map("${:,.2f}".format)
purchasing_analysis_table.head()

```

```python
purchase_df['Age Group'] = pd.cut(purchase_df['Age'], bins, labels=labels)
purchase_grouped = purchase_df.groupby('Age Group')
pv1 = purchase_grouped['Price'].sum()
pc1 = purchase_grouped['Price'].count()
ap1 = purchase_grouped['Price'].mean()
nt1 = pv1 / new_age_df['Total Count']
purchasing_analysis_age = pd.DataFrame({'Purchase Count': pc1, 'Average Purchase Price': ap1, 'Total Purchase Value': pv1, 'Normalized Totals': nt1})
purchasing_analysis_age['Average Purchase Price'] = purchasing_analysis_age['Average Purchase Price'].map("${:,.2f}".format)
purchasing_analysis_age['Normalized Totals'] = purchasing_analysis_age['Normalized Totals'].map("${:,.2f}".format)
purchasing_analysis_age['Total Purchase Value'] = purchasing_analysis_age['Total Purchase Value'].map("${:,.2f}".format)
purchasing_analysis_age.sort_index(inplace=True)
purchasing_analysis_age.head()
```
